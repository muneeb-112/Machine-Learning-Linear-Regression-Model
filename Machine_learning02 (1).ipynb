```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Student_Performance.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
def convert_text(extracurricular_activities):
  if extracurricular_activities == 'Yes': # Changed 'yes' to 'Yes' to match 'df.head()' output
    return 1
  else:
    return 0
df['Extracurricular Activities'] = df['Extracurricular Activities'].apply(convert_text)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Performance Index',axis=1)
y=df['Performance Index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(X_train, y_train)
```

```python
predicted_result = Lr.predict(X_test)
```

```python
sns.scatterplot(x=y_test, y=predicted_result)
```

```python
sns.displot(y_test-predicted_result)
```

```python
sns.displot(predicted_result)
```

```python
sns.boxplot(x=y_test, y=predicted_result)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, predicted_result)
```
